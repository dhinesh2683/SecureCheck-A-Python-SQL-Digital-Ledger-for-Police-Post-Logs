```python
import pandas as pd
from sqlalchemy import create_engine

# -----------------------------
# Step 1: Database Credentials
# -----------------------------
username = 'root'          # Your MySQL username
password = 'dhinesh'       # Your MySQL password
host = 'localhost'         # Database host
port = '3306'              # MySQL port (default: 3306)
database = 'traffic_logs'  # Ensure this database is created manually first

# -----------------------------
# Step 2: Connect to MySQL
# -----------------------------
try:
    engine = create_engine(f'mysql+mysqlconnector://{username}:{password}@{host}:{port}/{database}')
    print("✅ MySQL connection successful.")
except Exception as e:
    print("❌ Failed to connect to MySQL:", e)
    exit()

# -----------------------------
# Step 3: Load Excel File
# -----------------------------
try:
    df = pd.read_excel('cleaned_traffic_stops.xlsx')
    print(f"✅ Loaded Excel file. {len(df)} rows ready to insert.")
except Exception as e:
    print("❌ Failed to load Excel file:", e)
    exit()

# -----------------------------
# Step 4: Insert Data to MySQL
# -----------------------------
try:
    df.to_sql('traffic_stops', con=engine, if_exists='replace', index=False)
    print("✅ Data inserted successfully into 'traffic_stops' table.")
except Exception as e:
    print("❌ Failed to insert data into MySQL:", e)
```
